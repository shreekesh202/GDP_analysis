```python
df_ind=df.loc[(df['country_name']=='India') | (df['country_name']=="China, People's Republic of") | 
              (df['country_name']=="United States")]
df_ind.head()
```

```python
df_ind=df_ind.rename(index={77: 'India_Growth_Per_Year',36:'China_Yearly_Growth',186:'USA_Yearly_Growth'})
df_ind
```

```python
df_ind=df_ind.drop(['indicator_name','country_name'],axis=1)
df_ind
```

```python
df_ind=df_ind.transpose()
df_ind.index.name = 'Year'
df_ind.head()
```

```python
plt.figure(figsize=(10,7))
sns.lineplot(data=df_ind,x='Year',y='India_Growth_Per_Year',label='India')
sns.lineplot(data=df_ind,x='Year',y='China_Yearly_Growth',label='China')
sns.lineplot(data=df_ind,x='Year',y='USA_Yearly_Growth',label='USA')
plt.xticks(rotation=45)
plt.title('Yearly_growth_of_India_VS_China')
plt.show()
```
